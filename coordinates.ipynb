```python
import numpy as np
import pandas as pd
import re
geolocator = Nominatim(user_agent="hello")
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=5)
   
def get_location(row):
    address = row['besttelep']
    location = geocode(address)
    if location is None:
        address = re.sub(r'Municipiul','',address)
        address = re.sub(r'Comuna','',address)
        address = re.sub(r'Orasul','',address)
        location = geocode(address)
    row['longitude'] = location.longitude
    row['latitude'] = location.latitude
    print(address)
    return row

geolocated_table = telep_df.apply(get_location, axis=1)
```

```python
telep_df = df.merge(geolocated_table, on=['Telepules'])
```
